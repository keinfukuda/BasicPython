## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
csv_file_path2 = 'C:\\Users\\user\\OneDrive\\デスクトップ\\PG\\BasicPython\\data\\scatter_data.csv'
df_scatter_data = pd.read_csv(csv_file_path2)
df_scatter_data.head(10)
```

```python
df_class0 = df_scatter_data[df_scatter_data['class'] == 0].drop('class', axis=1)
df_class0.head()
```

```python
df_class0 = df_scatter_data[df_scatter_data['class'] == 0].drop('class', axis=1)
df_class1 = df_scatter_data[df_scatter_data['class'] == 1].drop('class', axis=1)
df_class2 = df_scatter_data[df_scatter_data['class'] == 2].drop('class', axis=1)
```

```python
plt.scatter(df_class0['length (cm)'], df_class0['width (cm)'], label='class0', color = "r")
plt.scatter(df_class1['length (cm)'], df_class1['width (cm)'], label='class1', color = "b")
plt.scatter(df_class2['length (cm)'], df_class2['width (cm)'], label='class2', color = "g")

plt.grid(True, linestyle='--', alpha=0.5)

plt.xlabel('length (cm)')
plt.ylabel('width (cm)')
plt.title('Scatter Plot of Three DataFrames')

plt.legend()
```

```python
csv_file_path3 = 'C:\\Users\\user\\OneDrive\\デスクトップ\\PG\\BasicPython\\data\\line_data.csv'
df_line_data = pd.read_csv(csv_file_path3)
df_line_data.head()
```

```python
plt.plot(df_line_data['期間'], df_line_data['件数'], marker='o', linestyle='-', color='blue', label='件数')

plt.grid(True, linestyle='--', alpha=0.5)

plt.xlabel('期間')
plt.ylabel('件数')
plt.title('期間ごとの件数の遷移')

plt.xticks(rotation=45)
```

```python
plt.bar(df_line_data['期間'], df_line_data['費用'], color='blue')

plt.grid(True, linestyle='--', alpha=0.5)

plt.xlabel('期間')
plt.ylabel('費用')
plt.title('期間ごとの費用の遷移')

plt.xticks(rotation=45)
```
